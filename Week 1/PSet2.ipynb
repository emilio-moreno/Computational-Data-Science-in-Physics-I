```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import sympy as sm
```

```python
# Error propagation
n = 10000
mux, muy, muz = [300000, 100000, 0.1]
sigmax, sigmay, sigmaz = [3000, 989, 0.01]
```

```python
def approximate_functional_std(function, variables = list, samples = list):
    sample_means = np.array([u.mean() for u in samples])
    sample_stds = np.array([u.std() for u in samples])
    
    # Partial derivatives evaluaated at mean
    substitution_list = np.array([(u, j) for u, j in zip(variables, sample_means)])
    partial_derivatives = np.array([function.diff(j).subs(substitution_list) for j in variables])
    
    # Approximation:
    approximated_variance = float(sum(partial_derivatives**2 * sample_stds**2))
    return np.sqrt(approximated_variance)
```

```python
# Variables
x, y, z = sm.symbols('x y z')
f = x**2 + y** + z**2 - sm.ln(x / y)
variables = [x, y, z]

# Samples
X = np.random.normal(mux, sigmax, n)
Y = np.random.normal(muy, sigmay, n)
Z = np.random.normal(muz, sigmaz, n)
samples = [X, Y, Z]

# Function evaluation
lam_f = sm.lambdify(variables, f)
W = lam_f(X, Y, Z)

print(f"Approximated STDV: {approximate_functional_std(f, [x, y, z], [X, Y, Z])}"
     f"\nSimulated STDV: {W.std()}")
```
